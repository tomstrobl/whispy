# Building block — Welcome & thank-you screens — configured by one file each

> **Building block.** A minimal, copy-pasteable block (config explanation +
> run) and nothing else. Paste it together with other blocks — e.g.
> `building_block_consent.ipynb` and a test block — to assemble your own
> experiment notebook.


This notebook shows the two **info screens** that frame an experiment: a
**welcome screen** at the very start and a **thank-you screen** after
everything is done. Both are plain `whispy.ui.InfoWindow`s and are
**config-driven**: each has one YAML file carrying a markdown `message` and a
`fullscreen` flag; the look (colors, fonts, button styling) comes from
[`configs/design.yml`](../configs/design.yml). You should not need to change
any Python below — just edit the YAML.

| file | shown | configures |
|---|---|---|
| [`configs/welcome.yml`](../configs/welcome.yml) | once, before anything else | `ui.message` (markdown), `ui.fullscreen` |
| [`configs/thanks.yml`](../configs/thanks.yml) | once, after the experiment | `ui.message` (markdown), `ui.fullscreen` |

Neither screen produces results, so there is nothing to save. Each cell
**blocks** until the participant clicks **Continue**.

If you haven't installed the project yet, run this from the repo root:
`pip install -e .`

## Welcome screen

Paste this at the **top** of your experiment notebook (before the consent
block). It greets the participant; reword the message in
[`configs/welcome.yml`](../configs/welcome.yml) (markdown supported).

```python
from pathlib import Path
from whispy.ui import InfoWindow
from whispy.utils import read_config

# The welcome message lives in configs/welcome.yml (markdown supported); reword
# it there without editing this cell.
welcome = read_config(Path('..') / 'configs' / 'welcome.yml')['ui']
InfoWindow(welcome['message'], fullscreen=welcome.get('fullscreen', True), blocking=True)
```

## Thank-you screen

Paste this at the **end** of your experiment notebook (after the last test
block). It tells the participant the experiment is over; reword the message in
[`configs/thanks.yml`](../configs/thanks.yml) (markdown supported).

```python
from pathlib import Path
from whispy.ui import InfoWindow
from whispy.utils import read_config

# The thank-you message lives in configs/thanks.yml (markdown supported); reword
# it there without editing this cell.
thanks = read_config(Path('..') / 'configs' / 'thanks.yml')['ui']
InfoWindow(thanks['message'], fullscreen=thanks.get('fullscreen', True), blocking=True)
```
